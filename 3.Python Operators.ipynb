# Python Operators
> ## Operators are used to perform operations on variables and values.

### Python divides the operators in the following groups:

* 1.Arithmetic operators
* 2.Assignment operators
* 3.Comparison operators
* 4.Logical operators
* 5.Identity operators
* 6.Membership operators
* 7.Bitwise operators

### 1.Python Arithmetic Operators

```python
#### Example for Python Arithmetic Operators
a = 5
b = 2

# addition
print ('Sum: ', a + b)  

# subtraction
print ('Subtraction: ', a - b)   

# multiplication
print ('Multiplication: ', a * b)  

# division
print ('Division: ', a / b) 

# modulo
print ('Modulo: ', a % b)  

# a to the power b (Exponentiation)
print ('Power: ', a ** b) 

# floor division
print ('Floor Division: ', a // b)
```

### 2. Python Assignment Operators

```python
### Examples
print('OUTPUTS : ')
print()
print('assign the sum of a and b to a')
# assign 10 to a
a = 10
# assign 5 to b
b = 5 
# assign the sum of a and b to a
a += b      # a = a + b
print(a)

print("assign the subtraction of a and b to a")

# assign 10 to a
a = 10
# assign 5 to b
b = 5 
# assign the subtraction of a and b to a
a -= b      # a = a + b
print(a)

print("assign the multipliclication of a and b to a")

# assign 10 to a
a = 10
# assign 5 to b
b = 5 
# assign the multiplication of a and b to a
a *= b      # a = a * b
print(a)

print('assign the divide of a and b to a')

# assign 10 to a
a = 10
# assign 5 to b
b = 5 
# assign the divide of a and b to a
a /= b      # a = a / b
print(a)

print('assign the modulus of a and b to a')

# assign 10 to a
a = 10
# assign 5 to b
b = 5 
# assign the modulus of a and b to a
a %= b      # a = a % b
print(a)

print('assign the floor division of a and b to a')

# assign 10 to a
a = 10
# assign 5 to b
b = 5 
# assign the floor division of a and b to a
a //= b      # a = a // b
print(a)

print('assign the power of a and b to a')

# assign 10 to a
a = 10
# assign 5 to b
b = 5 
# assign the power of a and b to a
a **= b      # a = a ** b
print(a)
```

### 3.Python Comparison Operators

```python
### Examples on comparison opertors 

a = 55

b = 25

# equal to operator
print('a == b =>', a == b)

# not equal to operator
print('a != b =>', a != b)

# greater than operator
print('a > b  =>', a > b)

# less than operator
print('a < b  =>', a < b)

# greater than or equal to operator
print('a >= b =>', a >= b)

# less than or equal to operator
print('a <= b =>', a <= b)
```

### 4.Python Logical Operators

```python
# logical AND
print('Logical AND output :')
print(True and True)     # True
print(True and False)    # False
print()
# logical OR
print('Logical OR output :')

print(True or False) # True
print(False or False) # False
print()
# logical NOT
print('Logical NOT output :')
print(not True)          # False
```

### 5.Python Identity Operators

```python
## Examples on Identify operators

a1 = 5
a2 = 5
b1 = 'Hello'
b2 = 'Hell0'
c1 = [1,2,3]
c2 = [1,2,3]

print(a1 is not b1)  # prints False

print(b1 is b2)  # prints True

print(c1 is c2)  #prints False  because they are equal but not identical. It is because the interpreter locates them separately in memory although they are equal.
```

### 6.Python Membership Operators

```python

x = 'vineeth reddy'
y = {1:'vin', 2:'red'}

# check if 'v' is present in x string
print('v' in x)  # prints True

# check if 'reddy' is present in x string
print('REDDY' not in x)  # prints True

# check if '1' key is present in y
print(1 in y)  # prints True

# check if 'a' key is present in y
print('v' in y)  # prints False
```

### 7.Python Bitwise Operators

```python

# bitwise operators
 
a = 100
b = 40
 
# Print bitwise AND operation
print("a & b =", a & b)
 
# Print bitwise OR operation
print("a | b =", a | b)
 
# Print bitwise NOT operation
print("~a =", ~a)
 
# print bitwise XOR operation
print("a ^ b =", a ^ b)

a = 10
b = -10
 
# print bitwise right shift operator
print("a >> 1 =", a >> 1)
print("b >> 1 =", b >> 1)
 
a = 5
b = -10
 
# print bitwise left shift operator
print("a << 1 =", a << 1)
print("b << 1 =", b << 1)
```
